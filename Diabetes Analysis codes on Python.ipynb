```python
import pandas as pd
```

```python
# Load the dataset
file_path = ("C:/Users/Boylikerichie/Documents/diabetes.csv")
df = pd.read_csv(file_path)
```

```python
# Display basic information and the first few rows
df.info(), df.head()
```

```python
import numpy as np

# Columns where zero values are likely incorrect
cols_to_clean = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]

# Replace zeros with NaN
df[cols_to_clean] = df[cols_to_clean].replace(0, np.nan)

# Check the number of missing values
df.isnull().sum()
```

```python
# Fill missing values with median of respective columns
df.fillna(df.median(), inplace=True)

# Verify if missing values are handled
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for seaborn
sns.set_style("whitegrid")

# Plot distribution of key numerical features
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

columns = ["Glucose", "BloodPressure", "BMI", "Age", "Insulin", "SkinThickness"]
for i, col in enumerate(columns):
    sns.histplot(df[col], bins=30, kde=True, ax=axes[i // 3, i % 3])
    axes[i // 3, i % 3].set_title(f"Distribution of {col}")

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df, hue="Outcome", diag_kind="kde", corner=True)
plt.show()
```

```python
# Boxplot: Comparing Distributions
sns.boxplot(x=df["Outcome"], y=df["Glucose"])
plt.title("Glucose Levels by Diabetes Outcome")
plt.show()
```

```python
#BMI Levels by Outcome
sns.boxplot(x=df["Outcome"], y=df["BMI"])
plt.title("BMI Distribution by Diabetes Outcome")
plt.show()
```

```python
# Glucose Levels by Outcome
sns.boxplot(x=df["Outcome"], y=df["Glucose"])
plt.title("Glucose Levels by Diabetes Outcome")
plt.show()
```

```python
# Countplot: How Many People Have Diabetes?
sns.countplot(x=df["Outcome"], palette=["green", "red"])
plt.title("Diabetes Cases Distribution")
plt.xticks([0, 1], ["No Diabetes", "Diabetes"])
plt.show()
```

```python
# Violin Plot:Distribution of Age by Outcome
sns.violinplot(x=df["Outcome"], y=df["Age"], split=True, inner="quartile")
plt.title("Age Distribution by Diabetes Outcome")
plt.show()
```

```python
# Heatmap: Correlation Between Features
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```
